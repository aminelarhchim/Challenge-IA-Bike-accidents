```python
# Read in the dataset
import pandas as pd

file_path = 'C:/Users/ladis/OneDrive/Bureau/carcteristiques-2021.csv'
df = pd.read_csv(file_path,";")

df = df.drop(['Num_Acc', 'jour', 'mois', 'an', 'hrmn', 'lum', 'dep', 'com', 'agg', 'int', 'atm', 'col', 'adr'], axis=1)

#print(df)
print(df.shape[0])

# Check if the dataset has at least 2 rows
if df.shape[0] < 2:
    print("Error: dataset needs at least 2 points to create a LineString")
else:
    # replace ',' with '.' in the lat and long columns
    df["lat"] = df["lat"].str.replace(',','.')
    df["long"] = df["long"].str.replace(',','.')
    #convert them to numeric
    df["lat"] = pd.to_numeric(df["lat"])
    df["long"] = pd.to_numeric(df["long"])
    print(df)
  

    # Print the dataframe
    print(df)
```
